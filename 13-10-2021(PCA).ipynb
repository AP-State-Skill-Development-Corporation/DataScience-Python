Dimensionality Reduction
- used to transform from high dimensional space to low dimensional space
- retain some meaningful properties from the original data

***Why Dimesionality Reduction***  

- higher dimensions leads to dimensionality curse
- if there are higher dimesions model performnace will be low
Use cases:
    - signal processing
    - speech recognition

***Tools for Dimensionality reduction***   

linear methods
- PCA-Principal Component Analysis
- LDA-Linear Discriminant Analysis   

Non Linear Methods  

- Kernel PCA
- Graph Based PCA
- GDA(generalised Discriminant Analysis)
- Auto encoders

#### Approches Dimensionality
- Feature Selection/Elimination
- Feature Extraction

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
print(cancer.DESCR)
```

```python
import pandas as pd
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
# Standardise the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
scaled_data=sc.fit_transform(df)
```

```python
scaled_data
```

```python
df1=pd.DataFrame(scaled_data,columns=cancer.feature_names)
df1.head()
```

```python
df.describe()
```

```python
### convert 30 features to 5features
from sklearn.decomposition import PCA
model=PCA(n_components=5)
```

```python
trans_data=model.fit_transform(df1)
```

```python
trans_data
```

```python
trans_data.shape
```

```python
trans_data
```

```python
Reduced_df=pd.DataFrame(trans_data,columns=["PC1",'PC2','PC3','PC4','PC5'])
Reduced_df["target"]=cancer.target
Reduced_df
```
