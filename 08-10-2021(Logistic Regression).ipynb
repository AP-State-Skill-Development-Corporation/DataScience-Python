```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Classification/Heart_disease.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Thal"].unique()
```

```python
import numpy as np
df["Thal"].replace(np.nan,"fixed",inplace=True)
df["Ca"].replace(np.nan,0.0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df["ChestPain"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df["ChestPain"]=label.fit_transform(df["ChestPain"])
df["Thal"]=label.fit_transform(df["Thal"])
```

```python
df
```

```python
df["Thal"].unique()
```

```python
x=df.drop(["AHD","Unnamed: 0"],axis=1)  #inputs/features
x
```

```python
y=df["AHD"]  #output/target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=25)
```

```python
x_train.shape
```

```python
df.shape
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### KNN Algorithm
K- no of nearest neighbors(sqrt(n))
1. calculate the euclidean distance
2. choose the no of nearest neighbors
3. compare with nearest neighbors
4. which got the majority vote will be select as final output category


```python
df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Classification/shirt.csv")
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df.drop("Size",axis=1)
```

```python
y=df["Size"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
nb=KNeighborsClassifier(n_neighbors=4)
```

```python
help(KNeighborsClassifier())
```

```python
nb.fit(x,y)
```

```python
y_pred=nb.predict(x)
```

```python
accuracy_score(y,y_pred)
```
