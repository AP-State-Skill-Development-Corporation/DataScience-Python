# Basic Plotting: Introduction to ```matplotlib```

In this section, we will:
- Create basic plots using ```matplotlib.pyplot```
- Put axis labels and titles
- Create multiple plots (subplots) in the same figure

```matplotlib``` is a python library. It contains the ```pyplot``` module, which is basically a collection of functions such as ```plot```, ```title```, ```show()``` etc. ```pyplot``` is one of the most commonly used module for creating a variety of plots such as line plots, bar plots, histograms etc.

## 1. Line plot
- Linear
- To generate single waves

````python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
````

````python
x = np.arange(5,40)
y = np.sqrt(x)
plt.plot(x,y)
plt.show()
````

````python
x = np.linspace(1,20,100)
y = np.cos(x)
plt.plot(x,y, marker = "+", linestyle = ":", color = "Black")
plt.show()
````

# Scatter Plot:
- Relation between two points
- to find out the probability
- Marketsing
- Cricket Graph
- share market

````python
x = np.random.randint(1,10, 10)
y = np.random.randint(1,10, 10)
plt.scatter(x,y, color = "r", marker = "P")
plt.show()
````

````python
help(plt.scatter)
````

````python
x = np.linspace(1,10,20)
y = np.power(x,2)
plt.scatter(x,y, color ="r", marker = "^")
plt.plot(x,y, color = "g")
plt.grid()
plt.show()
````

# Bar Graph:
- it is easy to understand by everyone
- corona cases in different cities
- Profits in stock market
- petrol rates in diffent countires
- vaccinations done in different cities



````python
# corona cases in different cities
covid_cases = [1200,2147,4524,2352,3545]
city_names = ["Vizag", "banglore"," Kurnool", "Vijayawada", "Tirupathi"] 
plt.bar(city_names,covid_cases, width = 0.7, color = ["r", "c", "g", "r", "b"])
plt.grid()
plt.text("Vizag",1200, "Low Level of Denger")
plt.xlabel("City Names in India")
plt.ylabel("No. of Covid Caeses of today")
plt.title("Covid Analysis")
plt.show()
````

````python
help(plt.bar)
````

````python
covid_cases = [1200,2147,4524,2352,3545]
city_names = ["Vizag", "banglore"," Kurnool", "Vijayawada", "Tirupathi"] 
plt.barh(city_names,covid_cases,  color = ["r", "c", "g", "r", "b"])
plt.grid()
plt.show()
````

# Histogram:
- Finding Probability

````python
x = [2,3,3,4,2,4,5,3,5,6,3,2,4,5,6,4,3,4,3]
plt.hist(x, histtype = "bar")
plt.grid()
plt.xlabel("Values")
plt.ylabel("Frequency count of our data")
plt.show()
````

````python
x = [2,3,3,4,2,4,5,3,5,6,3,2,4,5,6,4,3,4,3]
plt.hist(x, histtype = "step", align = "right")
plt.xlabel("Values",)
plt.ylabel("Frequency count of our data")
plt.show()
````

# Pie Chart
- population analycis

````python
marks = [90,69,45,73,29]
sub_names =  ["Python", "DataSciene", "C", "Machine Learning", "Java"]
plt.pie(marks, labels = sub_names, colors = ["r", "g", "m", "y", "c"],
        shadow= True, autopct  = "%.2f%%", radius = 1.5, explode = [0,0,0,0.2,0] )
plt.show()
````

````python
help(plt.pie)
````

````python
sns.catplot(x = "species", y = "sepal_length", data = iris)
plt.show()
````
