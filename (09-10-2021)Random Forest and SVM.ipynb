### Random Forest 
- ensembling classifier
- group of decision trees will be formed using random data,so it can be called as random forest
- we will consider the output of each decision tree and define the highest repeated times output as the predicted output

```python
import pandas as pd
df=pd.read_csv("C:/Users/prabhu/Downloads/archive(3)/diabetes.csv")
df
```

```python
df["Outcome"].unique()
```

```python
df.isna().sum()
```

```python
x=df.drop("Outcome",axis=1)
```

```python
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_t,x_te,y_t,y_te=train_test_split(x,y,train_size=0.9,random_state=1)
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_t,y_t)
```

```python
y_pred=model.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
print(accuracy_score(y_te,y_pred))
print(confusion_matrix(y_te,y_pred))
print(precision_score(y_te,y_pred))
print(recall_score(y_te,y_pred))
```
